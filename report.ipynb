## Section One: The Data Set

### The Original Data Set
#### Importing the Data and Describing the Data Set
The import data set is a csv file obtained from https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset. The owner of the data set is UCI Machine Learning and the data set was last updated about seven years ago. The data set is named "SMS Spam Collection Data Set" and it contains 5 columns and 5572 observations. The first column is named v1 and contains string values `spam` and `ham`. The second column is named v2 and contains string values denoting the sms message sent. The third, fourth, and fifth columns contain primarily null (NaN) values for most observations. 13.4% of the data set is spam messages and 86.6% of the data set is ham messages. 

```python
df = pd.read_csv("spam.csv", encoding='cp1252')
print("Number of Observations: ", len(df))
print("Number of Columns: ", len(df.columns))
print("Proportion of Spam Messages: ", len(df[df["v1"] == "spam"])/len(df))
print("Proportion of Ham Messages: ", len(df[df["v1"] == "ham"])/len(df))
df.head()
```

### Cleaning the Data & Feature Engineering
#### Replacing Unused Variables & Feature Engineering 
The first data frame, `spam_df`, will be a feature engineered version of the original data frame `df`. The new `spam_df` data frame will replace the `v1` and `v2` columns with the column names `Spam?` and `Message` respectively, and the `Spam?` column will now include integer value 0 if the message is `ham` and integer value 1 if the message is `spam`. As well, this data frame will drop the unnamed columns (`Unnamed: 2`, `Unnamed: 3`, and `Unnamed: 4`). 

Using feature engineering, new columns are added to the data frame to increase the number of factors influencing the results of the logistic regression and multinomial naive bayes tests. These new columns include `num_phone_related`, `num_winner`, `greetings`, `num_of_website`, `num_rand_letters`, `num_money`, `length_of_mess`, `spam_word_count`, and `ham_word_count`, which are all defined in the data preceding the displayed data frame. 

The `spam_word_count` and `ham_word_count` columns feature integer values counting the number of appearances of a 'spam' words and 'ham' words in each message. 'Spam' words are defined as the top thirty words that appear in spam messages in the data set. 'Ham' words are defined as the top thirty words that appear in ham messages in the data set. These words are listed in a table below the variable definitions.

```python
# Removing Excess Columns and Renaming
spam_df = df[["v1", "v2"]]
spam_df["Spam?"] = np.where(spam_df["v1"] == "spam", 1, 0)
spam_df = spam_df[["Spam?", "v2"]]
spam_df.rename(columns = {"v2": "Message"}, inplace=True)

# Feature Engineering New Variables
spam_df["num_phone_related"] = spam_df["Message"].str.lower().str.count("msg|sms|message|mobile|phone|call|txt|text|texting|send|reply") 
spam_df["num_winner"] = spam_df["Message"].str.lower().str.count("winner|congratulations|prize|jackpot|claim|vacation|cash|free|won|win|winning|iphone|money|ipad|cash") 
spam_df["greetings"] = spam_df["Message"].str.lower().str.count("hello|hi|what's up|whats up|sup|hey|heyy|heyyy|welcome|whatsup|hola|greetings|by|from")
spam_df["num_of_website"] = spam_df["Message"].str.lower().str.count("http|www|.com|.net|website|link|web|site|click|.gov") 
spam_df["num_rand_letters"] = spam_df["Message"].str.count("!|@|#|$|%|^|&|>|<|/+|=") - 2
spam_df["num_money"] = spam_df["Message"].str.lower().str.count("$|money|cash|card|free|credit|pay|venmo|cashapp|dollars|credits|dollar|cards") - 1
spam_df["length_of_mess"] = [len(mess) for mess in spam_df["Message"]]
spam_df["spam_word_count"] = spam_df["Message"].str.count("to|a|call|your|you|the|for|or|free|2|is|ur|on|txt|have|from|and|u|text|mobile|with|claim|reply|&|of|now|4|stop|our|this")
spam_df["ham_word_count"] = spam_df["Message"].str.count("i|you|to|the|a|u|and|in|my|is|me|of|for|that|it|have|your|but|are|so|not|on|at|i'm|can|if|do|will|be")
spam_df[:5]
```

#### Selecting the Thirty Most Common Spam and Ham Words
This barplot displays the thirty words that were found most often in messages labeled as spam in the data set, as well as the amount of times they appeared. As can be seen, some of the most common spam words that may play a roll in determining whether the message is spam or ham are 'call', 'free', 'txt', 'text', 'mobile', 'claim', and 'reply.'

```python
# Spam Most Common Words
spam_only = spam_df[spam_df["Spam?"] == 1]
spam_words = ""
for message in spam_only["Message"]:
    spam_words = spam_words.lower() + " " + message
spam_word_array = spam_words.split(" ")
spam_most_common = Counter(spam_word_array).most_common(30)
spam_words_df = pd.DataFrame(spam_most_common, columns = ["Word", "Count"])

# Barplot for Most Common Spam Words
sns.barplot(data=spam_words_df, x="Count", y="Word").set(title="Most Common Spam Words")
```

This barplot displays the thirty words that were found most often in messages labeled as spam in the data set, as well as the amount of times they appeared. As can be seen, some of the most common spam words that may play a roll in determining whether the message is spam or ham are 'can', 'have', 'u', 'but', 'if', 'do', and 'will.'

```python
# Ham Most Common Words 
ham_only = spam_df[spam_df["Spam?"] == 0]
ham_words = ""
for message in ham_only["Message"]:
    ham_words = ham_words.lower() + " " + message
ham_word_array = ham_words.split(" ")
ham_most_common = Counter(ham_word_array).most_common(30)
ham_words_df = pd.DataFrame(ham_most_common, columns = ["Word", "Count"])

# Barplot for Most Common Ham Words
sns.barplot(data=ham_words_df, x="Count", y="Word").set(title="Most Common Ham Words")
```

### Multinomial Naive Bayes Test
The next test I used was a Multinomial Naive Bayes algorithm. This model is most commonly use to determine the tag placed on a particular input. In this particular scenario, the multinomial naive bayes algorithm is determining the probability that each message is spam and the probability that each message is ham. From there, it selects the more probable case and labels the message as spam or ham. In order to calculate the probabilities, the algorithm is considering the variables that were feature engineered into the dataframe, such as number of websites and number of non-alphanumeric characters, defined in the [New Variable Definitions](#New-Variable-Definitions) section. As with the [Bernoulli Naive Bayes](#Bernoulli-Naive-Bayes-Test) algorithm, the Multinomial Naives Bayes algorithm is dependendent on the Naive Bayes probability theoreom show below:

$ P(A|B) = P(A) * \frac{P(B|A)}{P(B)} $

However, this formula shows Naive Bayes in its simplest form, where there is one independent variable and one dependent variable. In the case of this analysis, one possible scenario could be:

$A$: The Message is Spam

$B$: The Message contains 2 phone references, 1 winning reference, 1 greeting, 2 website references, 0 non-alphanumeric letters, 3 money references, 54 characters, 4 spam words, and 2 ham words

Therefore, the equation would be adapted to take into consideration all of the independent variables in order to find the probability of the message being, in this case, spam. This equation will replace B with the intersection of each independent variable, as seen below. 

$ P(A|X_1 \cap X_2 \cap ... \cap X_9) = P(A) * \frac{P(X_1 \cap X_2 \cap ... \cap X_9)|A)}{P(X_1 \cap X_2 
\cap ... \cap X_9)}$

with the variables $A$ and $X_i$ with $i$ being the i-th variable (in no particular order) representing

$A$ : The message is spam 

$X_1$ : 2 Phone References 

$X_2$: 1 Winning Reference 

... 

$X_9$ : 2 Ham Words 


The algorithm will calculate the probability of the message being spam and the probability of the message being ham given information about the feature engineered factors in the data frame. From there, it will select the possibility with a higher probability as its prediction. 

#### Performing the MNB Test
This sub-section features the code chunk that performs the Multinomial Naive Bayes classifier. It first calls the MultinomialNB() function imported from the Scikit-Learn package. Then, the function is fit with the training data (both the x and y values) allowing the learning algorithm to train itself and develop an understanding of which variables have high values for spam messages and which have high values for ham messages. Lastly, the algorithm uses what it has learned and makes predictions about the independent variable (x values) of the testing data set. These predictions are stored as a numpy array consisting of string values "Spam" or "Ham."

```python
# Test Two: Performing the MNB Test
mnb = MultinomialNB()
mnb.fit(X_train, y_train_target)
y_pred = mnb.predict(X_test)
```

### Bernoulli Naive Bayes Test
This next section dives into the third classifier: Bernoulli Naive Bayes. Bernoulli Naive Bayes, like Multinomial Naive Bayes, is based on the principles establishes by the Naive Bayes thereom. That being said, Bernoulli Naive Bayes will have the exact same probability formula as the Multinomial Naive Bayes classifier, shown below:

$ P(A|X_1 \cap X_2 \cap ... \cap X_9) = P(A) * \frac{P(X_1 \cap X_2 \cap ... \cap X_9)|A)}{P(X_1 \cap X_2 
\cap ... \cap X_9)}$

The main difference between the Multinomial Naive Bayes classifier and the Bernoulli Naive Bayes classifier is the type(s) of the independent variable(s). While they both can have several independent variables, the independent variables for the MNB classifier are occurrence counts whereas the independent variables for the BNB classifier are binary or boolean, similar to the dependent variable. For example, the variable `num_phone_related`, an integer counting the number of phone references, is no turned into the variable `Phone_Related` with type boolean. `Phone_Related` will be true if the `num_phone_related` is greater than 0, meaning that there is at least one reference to a phone-related word. Therefore, the Bernoulli Naive Bayes classifier is exactly the same as the Multinomial Naive Bayes but instead of having independent variables that are numerical, it is has independent variables that are binary categorical variables, containing value true or false. 
#### Preparing the BNB Dataframe
The code chunk below creates a new pandas dataframe using information contained in our original dataframe, spam_df. The `Spam?`, `Message`, and `Variable` variables carry over as they appear in the spam_df dataframe. As for the rest of the variables, excluding `length_of_mess`, `spam_word_count`, and `ham_word_count`, the code checks whether their values in the spam_df are greater than 0. If the value, or row, is greater than 0, then that row takes on a true value, meaning that the message is, for example, "phone-related". If the value, is less than or equal to 0, the row will take on a false value, meaning that the message is, for example, not "phone-related". Finally, the training and testing dataframes are prepared by dropping the `Spam?`, `Message`, and `Variable` columns, setting the target dataframe as the `Variable` column, and then using the train_test_split function of Scikit-Learn. Approximately 80% of the rows are placed within the training dataframe and the leftover 20% of row sare placed within the test dataframe (at a random state of 999). The first five rows of the training dataframe for independent variables are shown below. 
As mentioned, the `length_of_mess`, `spam_word_count`, and `ham_word_count` variables are either not included in the dataframe or are converted into a binary variable by a different means. `length_of_mess` is depicted in the dataframe as `Length_of_Mess` and is converted into a binary variable by checking if each input is greater than the mean length message. If the `length_of_mess` is greater than the mean of the `length_of_mess` column, then the `Length_of_Mess` column in the bernoulli dataframe will be true; otherwise, it will be false. This change was made when changing `length_of_mess` into a binary categorical variable as almost (if not, all) messages have a length greater than 0, so it wouldn't make much sense as all messages would have a true value for this column. `spam_word_count` and `ham_word_count` were both removed from the dataframe entirely for two reasons. First, they were hard to change into a categorical variable as there is a considerable amount of overlap between 'spam' words and 'ham' words and if we made out boundary 0, many of the values in each column would be true. Second, when implementing a similar method to the `length_of_mess` column using means, the classifier actually proved to have a lower accuracy score. That being said, it was best to remove these two variables from the bernoulli dataframe entirely. 

```python
# Test Three: Creating the Boolean Dataframe for BernoulliNB
bernoulli_df = pd.DataFrame({})
bernoulli_df["Spam?"] = spam_df["Spam?"]
bernoulli_df["Message"] = spam_df["Message"]
bernoulli_df["Variable"] = spam_df["Variable"]
bernoulli_df["Phone_Related"] = np.where(spam_df["num_phone_related"] > 0, True, False)
bernoulli_df["Winner_Related"] = np.where(spam_df["num_winner"] > 0, True, False)
bernoulli_df["Website_Related"] = np.where(spam_df["num_of_website"] > 0, True, False)
bernoulli_df["Random_Letters"] = np.where(spam_df["num_rand_letters"] > 0, True, False)
bernoulli_df["Greetings"] = np.where(spam_df["greetings"] > 0, True, False)
bernoulli_df["Money_Related"] = np.where(spam_df["num_money"] > 0, True, False)
bernoulli_df["Length_of_Mess"] = np.where(spam_df["length_of_mess"] > np.mean(spam_df["length_of_mess"]), True, False)

bernoulli_data = bernoulli_df.drop(columns=["Spam?", "Message", "Variable"])
bernoulli_target = bernoulli_df["Variable"]
X_train_bnb, X_test_bnb, y_train_target_bnb, y_test_target_bnb = train_test_split(bernoulli_data, bernoulli_target, test_size = 0.2, random_state=999)
X_train_bnb.head(20)
```

#### Performing the BNB Test
This subsection shows the preparation and execution of the Bernoulli Naive Bayes algorithm. First, the code chunk introduces the Bernoulli Naive Bayes classifier in the variable `bnb`. From there, it fits the classifier to both the training independent variables and the training dependent variables using the fit function of Scikit-Learn BernoulliNB package. In doing so, the classifier is able to train itself on the training data to establish and recognize patterns that will allow it to more accurately make predictions on the testing dataframe independent variables. Lastly, the code chunk makes these predictions on the testing dataframe and saves it in the variable `y_pred`.

```python
# Test Three: Performing the BNB Test
bnb = BernoulliNB()
bnb.fit(X_train_bnb, y_train_target_bnb)
y_pred = bnb.predict(X_test_bnb)
```

### Using Count Vectorizer for MNB
A count vectorizer differs greatly from the dataframes we've used in the previous three sub-sections. The Count Vectorizer is a Scikit-Learn tool that takes each word in a given message and creates a column for each unique word, counting the number of times that word appears in the message. It repeats this process for every message in the data frame. Therefore, if a certain word in a message does not have a column after iterating through the words of the previous messages, it creates a new column for that word, placing zeros in columns where the word doesn't appear, and counting the number of appearances in messages in which it does appear. In simpler terms, each word in a dataframe column (`Message`) has a column and it counts the number of appearances of each word in each message. As well, it defaults to converting each word to lowercase, allowing us to account for awkward casing. An example is shown below:

`Message`: "I love to RuN walk and to lOvE everyone I Run with"

|  | `i` | `love` | `to` | `run` | `walk` | `and` | `everyone` | `with` |
|---|---|---|---|---|---|---|---|---|
| 0 | 2 | 2 | 2 | 2 | 1 | 1 | 1 | 1| 


For the previous algorithms, we used a hand-crafted feature-engineered dataframe classifying messages under specific topics typically found in spam dataframes. With this feature-engineer dataframe, we split it into training and testing dataframes and fit the training dataframe to a logistic regression, a Multinomial Naive Bayes classifier, and then a Bernoulli Naive Bayes classifier. In this section, we are using a new dataframe crafted by the Count Vectorizer function of scikit-learn that is explained in the text above. Using this count vectorizer, we could evaluate how much more or less accurate using a count vectorizer is in comparison to hand feature-engineering dataframes for each of these classifiers. 
#### Preparing the CV Dataframe
In order to prepare the dataframe for the count vectorizer, all that is needed is the message and its classification as spam or ham. Therefore, the independent variable is `Message` and dependent variable is `Variable` (or `Spam?` in words), which is split into a training and testing data frame. Using the scikit-learn train_test_split function, these independent and dependent variable dataframes are split with 80% of the messages making up the training dataframes and 20% making up the testing dataframes (with a random state of 999). First, the code introduces the Count Vectorizer of the scikit-learn package as count_vect. Then, it fits the independent variables training dataframe to the count vectorizer to count the appearance of each unique word in the dataframe for each message, similar to the example shown in the [Count Vectorizer Section](#Count-Vectorizer). One particular input for the count vectorizer function is a choice of stop words, meaning a group of words that the count vectorizer counts from. A typical value for `stop_words` is 'english', declaring that the count vectorizer will only count words in the message that are English. Read the [next section](#Performing-the-MNB-Test-on-the-CV-Dataframe) to see how the transformed dataframe is fit to the Multinomial Naive Bayes classifier.

We used seaborn to identify the most common words in the dataframe and show the quantity of each of these words.

In section two, we used the seaborn package in python to visualize eight of these variables. Using the barplot function of seaborn, we demonstrated the difference in means for spam and ham messages for each of these variables. For example, the barplot in the section [Mean Number of Money References for Spam and Ham](#Visualization-Six:-Mean-Number-of-Money-References-for-Spam-and-Ham) shows two bars, each representing the mean number of money references for either spam or ham messages. These visualizations helped us understand the significance of certain variables that were feature engineered into the dataframe. For example, the large gap in mean number of money references between spam and ham messages may indicate that this is a good variable to include in the model. The final visualization was a scatterplot showing the number of spam words vs the number of ham words in each message, while also indicating whether each point (representing a message) was spam or ham. This allowed us to conclude, as expected, that spam messages typically had more spam words than ham message did, and vice versa. 

Finally, in section three, we applied machine learning skills on a training dataframe to make predictions about a test dataframe. This first machine learning model that was implemented as a Logistic Regression on the hand-feature-engineered dataframe, which yielded an accuracy score of 95.07%. The second machine learning classifier we used was Multinomial Naive Bayes classifier on the same dataframe as the Logistic Regression. This classifier had the worst performance, with an accuracy score of only 92.56%. Next, we used a Bernoulli Naive Bayes classifier on an altered dataframe. This yielded an accuracy score of 93.18%, which is slightly better than that of the Multinomial Naive Bayes classifier, but less than that of the Logistic Regression. For the fourth machine learning model, we used a count vectorizer to count the appearances of every word in the dataframe in each message.
